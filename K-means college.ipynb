# K-means Cluster Formation

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df.drop('Private',axis=1))
```

```python
#cluster centre vectors(dimensions will match the n_features)
```

```python
kmeans.cluster_centers_
```

# Evaluation

```python
# we are converting the private column to basically dummy variables using apply function
```

```python
def converter(private):
    if private == 'Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
df.head()
```

```python
#creating confusion matrix and classification report to see how well the clustering worked without being any levels
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(df['Cluster'],kmeans.labels_))
```

```python
print(classification_report(df['Cluster'],kmeans.labels_))
```

```python
#the algorithm is working pretty well .. it is using the features to distint 2 groups
```
